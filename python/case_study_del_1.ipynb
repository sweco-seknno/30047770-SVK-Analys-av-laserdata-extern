## Installation av nödvändig programvara och Python-paket
*Tidsuppskattning: 35 minuter*


**PostgreSQL** behöver installeras med tillägg för PostGIS.

1. Ladda ner via [installation av PostgreSQL och PostGIS](https://www.enterprisedb.com/downloads/postgres-postgresql-downloads). [Installations-instruktioner](https://www.bostongis.com/PrinterFriendly.aspx?content_name=postgis_tut01). Under steget *Password* väljer du ett lösenord för superanvändaren postgres. Skriv gärna ner lösenordet så att du kommer ihåg det.
2. Under steget *Välj applikationer att installera* klicka i *Spatial Extensions* > *PostGIS...*. Om inte detta val finns, gå till steg 3. 
3. Om steg 2 inte fungerade: [Installera PostGIS manuellt](https://winnie.postgis.net/download/windows/). Öppna mappen motsvarande den version av PostgreSQL som du laddade ner, t.ex.  *pg17*. Öppna mappen *buildbot*. Ladda ner filen med ändelsen exe, t.ex. *postgis-bundle-pg17x64-setup-3.5.2-1.exe*.
4. Öppna cmd terminal. Byt ut 17 mot versionen av PostgreSQL som du laddade ner i samtliga efterföljande steg:

    ```
    cd C:\Program Files\PostgreSQL\17\bin
    ```

5. Initialisera PostgreSQL i cmd terminal:

    ```
    initdb -D "C:\Program Files\PostgreSQL\17\data" -U postgres -A scram-sha-256 -W
    ```

6. Starta psql-server:

    ```
    pg_ctl -D "C:\Program Files\PostgreSQL\17\data" start
    ```

    **Tips:** När du vid ett senare tillfälle vill stoppa psql-servern:

    ```
    pg_ctl -D "C:\Program Files\PostgreSQL\17\data" stop
    ```

**ogr2ogr** är ett kommandoradsverktyg från GDAL som kan användas för att konvertera data mellan formaten PostGIS och Esri filgeodatabas.

1. Installera OSGeo4W som inkluderar en installation av GDAL. [Följ instruktionerna under *Quick Start for OSGeo4W Users*](https://trac.osgeo.org/osgeo4w/).
2. Klicka i *GDAL* under steget *Select Packages*.
3. Öppna *OSGeo4W Shell* när installationen är klar (sök efter det genom startmenyn).
4. Verifiera installationen genom kommandot:

    ```
    ogr2ogr --version
    ```

**psycopg** är ett paket som krävs för att kunna koppla upp mot och hantera PostGIS-databaser i Python.

1. Installera paketet till projektets virtuella miljö:

    ```
    pip install psycopg
    ```
